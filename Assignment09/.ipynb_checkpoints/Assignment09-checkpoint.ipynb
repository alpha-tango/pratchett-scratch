```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
from scipy.special import zeta
from statistics import mean 
import statsmodels.api as sm
```

# Assignment 9

## Question 1

### Part A

For Î³ = 5/2, generate n = 1000 sets each of N = 10, 10^2, 10^3, 10^4, 10^5, and 10^6
samples, using Pk = ck^5/2 with k = 1, 2, 3, . . .

https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.powerlaw.html -> only handles exponents > 0

```python
gamma = 5/2
n = np.arange(1, 1000 + 1)
N = [10, 10**2, 10**3, 10**4, 10**5, 10**6]
```

```python
k_values = np.arange(1, 10**6 + 2)

f = pd.DataFrame(k_values, columns=['k'])
# f['p'] = f['k'] ** -gamma
f['p'] = 1 / zeta(gamma) * f['k'] ** -gamma
total = f['p'].sum()
f['discrete_pdf'] = f['p'] / total
f = f.sort_values(by='k')  # should be unnecessary but let's be careful
f['discrete_cdf'] = f['discrete_pdf'].cumsum()
f.tail(20) 
```

```python
f.head()
```

```python
plt.plot(np.log10(f['k']), np.log10(f['discrete_cdf']))
```

```python
x = random.random()
print(x)
```

```python
sample_k = f[f['discrete_cdf'] >= x].min()['k']
```

```python
k_max_values = []

for sample_size in N:
    N_k_max_values = []
    
    for i in n:  # 1000 sets of samples
                
        samples = []
        for j in range(sample_size):  # generate N samples
            x = random.random() 
            sample_k = int((1 - x)**(- 2/3))  # Given x is distributed uniformly on [0, 1], then k = [ (1- x)** - 2/3]
            # sample_k = f[f['discrete_cdf'] >= x].min()['k']
            samples.append(sample_k)  
            
            
        k_max_i = max(samples)   # take the max of all the random numbers in that sample set
        N_k_max_values.append(k_max_i)  # save it
    
    k_max_values.append(N_k_max_values)
```

```python
fig = plt.figure(figsize=(20,18))

total_plots = len(N)
cols = 2
rows = -(total_plots // -cols)  # ceiling division shorthand
position = range(1, total_plots + 1)

for k in range(total_plots):
    ax = fig.add_subplot(rows, cols, position[k])
    ax.set_title(f"N = {N[k]}")
    ax.set_xlabel("n")
    ax.set_ylabel("log10 k_max_i")
    ax.set_ylim(bottom=0.0, top=7.5)
    x = n
    y = np.log10(k_max_values[k])
    ax.scatter(x, y)
```

### Part B

```python
avg_k_values = []

for i in range(len(N)):
    avg_k_values.append(mean(k_max_values[i]))
```

```python
x  = np.log10(N)
y = np.log10(avg_k_values)

X = sm.add_constant(x)
line_1 = sm.OLS(y, X).fit()

slope_1 = line_1.params[1]
intercept_1 = line_1.params[0]

y_fit = intercept_1 + x * slope_1

print(line_1.summary2())
```

```python
figb, axb = plt.subplots()
axb.set_xlabel("log10 N")
axb.set_ylabel("log10 â¨kmaxâ©")
axb.scatter(x, y)
axb.text(2,4, s=f"slope = 0.68")
axb.plot(x, y_fit)
```
