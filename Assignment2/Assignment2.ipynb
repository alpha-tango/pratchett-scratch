```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read in the data
world_record_csv = "row_times.csv"
world_records = pd.read_csv(world_record_csv)
world_records.head()
```

```python
world_records.info()
```

```python
world_records['timedelta'] = pd.to_timedelta("00:" + world_records['time']) # default in nanoseconds
world_records['km_p_h'] = 2 / (world_records['timedelta'].dt.seconds / 60 / 60)
world_records['log10_oarsfolk'] = np.log10(world_records['no_oarsfolk'])
world_records['log10_kmph'] = np.log10(world_records['km_p_h'])
```

```python
world_records.head()
```

```python
world_records.info()
```

```python
# Plot data
x = world_records['no_oarsfolk']
y = world_records['km_p_h']
plt.xlabel('Number of Oarsfolk')
plt.ylabel('km/h')
plt.scatter(x, y) 

# Include intercept
plt.xlim(xmin=0)
plt.ylim(ymin=0)
plt.show()
```

```python
def fit_and_plot(plt, x, y, label):
    # get fit line
    slope, intercept = np.polyfit(x, y, 1)

    # Write slope
    scaling_text = f" (scaling exponent = {slope:.2f})"
    l = label + scaling_text
    print(l)
    
    # Plot data
    plt.scatter(x, y, label=(label + scaling_text))
    
    # Add fit line
    plt.plot(x, slope*x+intercept) 
    
    # Include intercept
    plt.xlim(xmin=0, xmax=1.0)
    plt.ylim(ymin=0, ymax=1.5)
    

    # plt.text(0.1, 0.5, scaling_text, horizontalalignment='left', verticalalignment='center')

    # Label series
    
    
    return(plt)
# set up axes
x = world_records['log10_oarsfolk']
y = world_records['log10_kmph']
plt.xlabel('Number of Oarsfolk')
plt.ylabel('km/h')

plt = fit_and_plot(plt, x,y, label='all data')
plt.legend(loc='lower right')
plt.show()
```

```python
# set up new series
x_women = world_records[(world_records['gender'] == 'women')]['log10_oarsfolk']
y_women = world_records[(world_records['gender'] == 'women')]['log10_kmph']

x_men = world_records[(world_records['gender'] == 'men')]['log10_oarsfolk']
y_men = world_records[(world_records['gender'] == 'men')]['log10_kmph']

plt.xlabel('Number of Oarsfolk')
plt.ylabel('km/h')

plt = fit_and_plot(plt, x_men,y_men, label='men')
plt = fit_and_plot(plt, x_women,y_women, label='women')
plt.legend(loc='lower right')
plt.show()
```

```python
# set up new series
x_1oar = world_records[(world_records['oars_per_rower'] == 1)]['log10_oarsfolk']
y_1oar = world_records[(world_records['oars_per_rower'] == 1)]['log10_kmph']

x_2oar = world_records[(world_records['oars_per_rower'] == 2)]['log10_oarsfolk']
y_2oar = world_records[(world_records['oars_per_rower'] == 2)]['log10_kmph']

plt.xlabel('Number of Oarsfolk')
plt.ylabel('km/h')

plt = fit_and_plot(plt, x_1oar,y_1oar, label='1 oar each')
plt = fit_and_plot(plt, x_2oar,y_2oar, label='2 oars each')
plt.legend(loc='lower right')
plt.show()
```

```python
# set up new series
x_coxed = world_records[(world_records['coxed'] == True)]['log10_oarsfolk']
y_coxed = world_records[(world_records['coxed'] == True)]['log10_kmph']

x_uncoxed = world_records[(world_records['coxed'] == False)]['log10_oarsfolk']
y_uncoxed = world_records[(world_records['coxed'] == False)]['log10_kmph']

plt.xlabel('Number of Oarsfolk')
plt.ylabel('km/h')

plt = fit_and_plot(plt, x_coxed,y_coxed, label='coxed')
plt = fit_and_plot(plt, x_uncoxed,y_uncoxed, label='uncoxed')
plt.legend(loc='lower right')
plt.show()
```

```python
# set up new series
x_norm = world_records[(world_records['lightweight'] == False)]['log10_oarsfolk']
y_norm = world_records[(world_records['lightweight'] == False)]['log10_kmph']

x_lightweight = world_records[(world_records['lightweight'] == True)]['log10_oarsfolk']
y_lightweight = world_records[(world_records['lightweight'] == True)]['log10_kmph']

plt.xlabel('Number of Oarsfolk')
plt.ylabel('km/h')

plt = fit_and_plot(plt, x_norm,y_norm, label='general')
plt = fit_and_plot(plt, x_lightweight,y_lightweight, label='lightweight')
plt.legend(loc='lower right')
plt.show()
```

```python
## load planet data
planets_csv = "planets.csv"
planets = pd.read_csv(planets_csv)
planets.head()
```

```python
planets['radius_km'] = planets['radius_million_km'] * 10**6
planets['tau_days_sq'] = planets['tau_days'] ** 2
planets['radius_km_cube'] = planets['radius_km'] ** 3
planets['log10_tau_days_sq'] = np.log10(planets['tau_days_sq'])
planets['log10_radius_km_cube'] = np.log10(planets['radius_km_cube'])
planets.head(10)
```

```python
# set up axes
x = planets['tau_days_sq']
y = planets['radius_km_cube']
plt.xlabel('Tau (days) ^2')
plt.ylabel('Radius (km) ^3')

slope, intercept = np.polyfit(x, y, 1)

# Write slope
scaling_text = f" (scaling exponent = {slope:.2f})"
print(scaling_text)

# Plot data
plt.scatter(x, y)

# Add fit line
plt.plot(x, slope*x+intercept) 

# Mark points with planet names
for (xi, yi, y_label) in zip(x, y, planets['planet']):
    print(xi, yi, y_label)
    plt.text(xi, yi, y_label, va='bottom', ha='center')

# write slope
plt.text(0.1, 0.5, scaling_text, horizontalalignment='left', verticalalignment='center')
plt.show()
```

```python
# set up axes
x = planets['log10_tau_days_sq']
y = planets['log10_radius_km_cube']
plt.xlabel('Log10 (Tau (days) ^2)')
plt.ylabel('Log10 (Radius (km) ^3)')

slope, intercept = np.polyfit(x, y, 1)

# Write slope
scaling_text = f"scaling exponent = {slope:.2f}"
print(scaling_text)

# Plot data
plt.scatter(x, y)

# Add fit line
plt.plot(x, slope*x+intercept) 
plt.text(7, 24, scaling_text, horizontalalignment='left', verticalalignment='center')

# Mark points with planet names
for (xi, yi, y_label) in zip(x, y, planets['planet']):
    plt.text(xi, yi, y_label, va='bottom', ha='center')

plt.show()
```
